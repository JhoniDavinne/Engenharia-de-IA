Desafio:
1. Carregar um arquivo .csv com feedback de clientes (reviews.csv)
2. Usar um LLM para classificar o sentimento de cada feedback (Positivo, Negativo, Neutro)
3. Adicionar uma nova coluna com essa classificação ao DataFrame.

````python
# Etapa 1
df_reviews = pd.read_csv("/content/reviews.csv")
````

````python
df_reviews.head()
````

````python
coluna_de_reviews = df_reviews["reviewText"]
coluna_de_reviews
````

### Few-shot prompting

````python
lista_de_analises_de_sentimentos = []

for review_numero, resenha in enumerate(coluna_de_reviews):
    resposta = client_groq.chat.completions.create(
        model="llama-3.1-8b-instant", # Updated to a newer, likely supported Groq model
        messages=[
            {
                "role": "user",
                "content": f"""Você irá analisar a resenha que eu te mandarei abaixo, e retorna com uma análise de sentimento.
                    Você deve responder APENAS com uma das seguintes palavras: 'Positiva', 'Negativa' ou 'Neutra',
                    indicando o sentimento relativo àquela resenha específica.

                    Exemplos:
                    'Eu adorei esse produto' -> Positiva
                    'Gostei, mas não é nada de especial' -> Neutra
                    'Odiei esse produto' -> Negativa

                    Segue a resenha a ser analisada: {resenha}
                    """,
            }
        ],
        temperature=0 # Setting temperature to 0 for consistent responses
    )
    lista_de_analises_de_sentimentos.append(resposta.choices[0].message.content)
    print(f"Resenha {review_numero}: '{resenha}' -> Sentimento: {resposta.choices[0].message.content}")
````

````python
lista_de_analises_de_sentimentos
````

````python
# Etapa 3
df_reviews["Análises de Sentimentos"] = lista_de_analises_de_sentimentos
````

````python
df_reviews
````

````python
df_reviews_negativas = df_reviews[df_reviews["Análises de Sentimentos"] == "Negativa"]
df_reviews_negativas
````

````python
resenhas_negativas = df_reviews_negativas["reviewText"]
resenhas_negativas
````

````python
resenhas_negativas_unidas = "#####".join(resenhas_negativas)
resenhas_negativas_unidas
````

````python
full_summary = ""
resposta_de_categoria = client_groq.chat.completions.create(
    model="llama-3.1-8b-instant", # Changed to a Groq-supported model
    messages=[
        {
            "role": "user",
            "content": f""" Você é um analista de dados. Vou te passa muitas resenhas negativas de
            análises de um produto, separadas por '#####', e eu quero que você encontre 5 categorias diferentes para os tipos de reclamações.
            Quero que você me retorne 5 categorias.
            Cada categoria deve ser definida por APENAS uma palavra.
            Não escreva nada mais além das 5 categorias. Quero que elas sejam retornadas
            separadas por virgula e em letra minúscula, e sem acetos gráficos.

            exemplo:

            'eletronicas, roupas, alimentos, viagens, brinquedos'

            Aqui estão as resenhas negativas: {resenhas_negativas_unidas}
                """,
        }
    ],
    temperature=0, # Setting temperature to 0 for consistent responses
    stream=True # Added stream=True for iteration
)

for chunk in resposta_de_categoria:
  if chunk.choices[0].delta.content:
      full_summary += chunk.choices[0].delta.content

print(full_summary)
````

````python
lista_de_caterias = full_summary.split(", ")
lista_de_caterias
````

````python
import os
from google.colab import userdata

os.environ["GOOGLE_API_KEY"] = userdata.get('GEMINI_API_KEY')
from google import genai

client = genai.Client()
````

````python
dicionariao_de_resennhas_classificadas = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=f""" Você é um analista de dados. Vou te passa muitas resenhas negativas de
            análises de um produto, separadas por '#####', e eu quero que você encontre 5 categorias diferentes
            para os tipos de reclamações. Cada categoria deve ser definida por APENAS uma palavra.

            Depois, eu quero quer você me retorne APENAS um texto em formato JSON, contendo
            três chaves:
            - 'resenha_original' irá conter a resenha original em inglês.
            - 'resenha_pt' irá conter a resenha traduzida para o português.
            - 'categoria' irá conter a categoria dentre as 5 definidas por você em que a resenha se encaixa.

            Exemplo de como deve ser a estrutura, a ser seguida:

            {{'resenha_original': 'I didn't like the color of product',
              'resenha_pt': 'Eu não gostei da cor do produto',
              'categoria': 'design'}}

            Retorne APENAS o texto JSON. Não escreva nada mais além do texto JSON.

            Não retorne nada além do JSON, NÃO escreva nada além do JSON.
            RETONE SOMENTE O JSON.

            Faça isso para todas as resenhas que eu vou te enviar abaixo: {resenhas_negativas_unidas}
            """
)
json_de_resenhas_classificadas = dicionariao_de_resennhas_classificadas.text
print(json_de_resenhas_classificadas)
````
